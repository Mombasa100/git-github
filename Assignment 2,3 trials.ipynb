```python
# imports needed to complete the assignment

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## import file

```python
data = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

## Breakdown of distance from home by job role and attrition

```python
data.groupby(['DistanceFromHome']).sum()
```

```python
data.groupby(['JobRole']).sum()
```

```python
data.groupby(['Attrition']).sum()
```

## Comparison between monthly income by education and attrition

```python
data.groupby(["Education", "Attrition"])["MonthlyIncome"].sum()
```

###### Distribution of employees by distance from home,job role,attrition,education

```python
data.DistanceFromHome.value_counts().plot.bar(figsize=(15,8), color = 'teal')
plt.title(' Distance From Home')
plt.xlabel('Job role')
plt.ylabel('Attrition')
plt.show()
```

#### explore attrition by age

```python
plt.figure(figsize = (10,8))
sns.countplot(x = data.Age, hue = data.Attrition)
plt.title('Attrition by Age')
plt.show()
```

### data for education

```python
data.Education.value_counts()
```

```python
data.Education.value_counts().plot.barh(figsize=(10,6))
plt.xlabel('Number of employees')
plt.ylabel('Education')
plt.title('Education of Employees')
plt.show()
```

```python
data.Education.value_counts().plot.pie(figsize=(10,6), startangle = 5, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
